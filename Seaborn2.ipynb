# 1. Pre√ßos outliers de diamante

```python
import seaborn as sns

data = sns.load_dataset('diamonds')
data.head()
```

```python
with sns.axes_style('whitegrid'):
  diamonds = sns.boxplot(x=data['price'])
  diamonds.set(title='Distribui√ß√£o de pre√ßos', xlabel='Pre√ßo');
```

```python
# Os valores se concentram em torno de 1000 e 5000 mil, por√©m existem alguns valores pequenos ate 12500, e depois outra pequena concentra√ß√£o de pre√ßos acima de 12500 at√© a acima de 17500
```

# 2. Correla√ß√£o entre o do pre√ßo do diamante com seu peso (`carat`) agrupados por sua transpar√™cia (`clarity`)

```python
with sns.axes_style('whitegrid'):
  diamonds2 = data.groupby('clarity')
  diamonds2 = sns.histplot(x=data['price'], y=data['carat'])
  diamonds2.set(title='Correla√ß√£o entre o do pre√ßo do diamante com seu peso `carat` agrupados por sua transpar√™cia `clarity`', xlabel='Pre√ßo', ylabel='Peso');
```

```python
with sns.axes_style('whitegrid'):

  diamond = sns.scatterplot(data=data, x='price', y='carat', hue='clarity', palette='pastel')
  diamond.set(title='Correla√ß√£o entre o do pre√ßo do diamante com seu peso `carat` agrupados por sua transpar√™cia `clarity`', xlabel='Pre√ßo', ylabel='Peso');
  diamond.get_legend().set_title('Transpar√™ncia');
```

```python
# Existe uma correla√ß√£o entre o peso e o pre√ßo do diamante, n√£o obstante a transpar√™ncia tamb√©m influ√™ncia no valor.
```
