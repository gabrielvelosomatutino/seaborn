# 1. Preços outliers de diamante

```python
import seaborn as sns

data = sns.load_dataset('diamonds')
data.head()
```

```python
with sns.axes_style('whitegrid'):
  diamonds = sns.boxplot(x=data['price'])
  diamonds.set(title='Distribuição de preços', xlabel='Preço');
```

```python
# Os valores se concentram em torno de 1000 e 5000 mil, porém existem alguns valores pequenos ate 12500, e depois outra pequena concentração de preços acima de 12500 até a acima de 17500
```

# 2. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```python
with sns.axes_style('whitegrid'):
  diamonds2 = data.groupby('clarity')
  diamonds2 = sns.histplot(x=data['price'], y=data['carat'])
  diamonds2.set(title='Correlação entre o do preço do diamante com seu peso `carat` agrupados por sua transparêcia `clarity`', xlabel='Preço', ylabel='Peso');
```

```python
with sns.axes_style('whitegrid'):

  diamond = sns.scatterplot(data=data, x='price', y='carat', hue='clarity', palette='pastel')
  diamond.set(title='Correlação entre o do preço do diamante com seu peso `carat` agrupados por sua transparêcia `clarity`', xlabel='Preço', ylabel='Peso');
  diamond.get_legend().set_title('Transparência');
```

```python
# Existe uma correlação entre o peso e o preço do diamante, não obstante a transparência também influência no valor.
```
