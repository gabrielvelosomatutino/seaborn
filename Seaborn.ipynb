# 1. Pre√ßo do diamante por tipo de corte

```python
import seaborn as sns
```

```python
data = sns.load_dataset('diamonds')
data.head()
```

```python
grafico1 = sns.barplot(data=data,x="cut", y="price")
grafico1.set(title='Pre√ßo do diamante por tipo de corte', xlabel='corte', ylabel='pre√ßo')
```

```python
# Com o gr√°fico pode-se perceber que h√° uma pequena vari√¢ncia no pre√ßo de acordo com o tipo de corte, todavia existem outras vari√°veis que podem afetar o pre√ßo em conjunto com o corte.
```

# 2. N√∫mero de passageiros em dezembro por ano

```python
import seaborn as sns
data2 = sns.load_dataset('flights')
data2.head()
```

```python
flight = data2.loc[data2['month']=='Dec']
grafico2 = sns.barplot(data=flight, x='year', y='passengers')
grafico2.set(title='N√∫mero de passageiros em dezembro por ano', xlabel='Ano', ylabel='Passageiros')
```

```python
# Com o gr√°fico √© poss√≠vel verificar que houve um crescimento gradual do n√∫mero de passageiros no m√™s de Dezembro a cada ano.
```

# 3. Numero de Passageiros por m√™s entre 1949 e 1959

```python
import seaborn as sns
data3 = sns.load_dataset('flights')
data3.head()
```

```python
flights2 = data3.query("1949 <= year < 1960")

with sns.axes_style('whitegrid'):

  grafico3 = sns.lineplot(data=flights2,x="month", y="passengers", hue="year", palette='pastel')
  grafico3.set(title='Numero de Passageiros por m√™s entre 1949 e 1959', xlabel='M√™s', ylabel='Passageiros');
  grafico3.get_legend().set_title('Ano');
```

```python
# Pode-se verificar pelo gr√°fico que h√° um aumento de passageiros gradual em decorrer dos anos, todavia h√° um aumento a cada m√™s dentro de cada ano, excepcionalmente entre osmeses de maio e setembro.
```
